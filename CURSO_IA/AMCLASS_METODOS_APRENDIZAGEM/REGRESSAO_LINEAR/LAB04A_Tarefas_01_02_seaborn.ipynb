## LAB 04A — Tarefa 01

Base de dados: `tips`

Essa base contém dados relacionados a contas de restaurante, incluindo valor total da conta, gorjeta, sexo, fumante ou não, dia, horário e tamanho da mesa.

Nesta tarefa, a variável que será usada como saída é:

- `tip`: valor da gorjeta.

```python
# LAB 04A 
# TAREFA 01

# Importa a biblioteca seaborn.
# O seaborn possui bases de dados prontas para estudo e também ferramentas para gráficos estatísticos.
import seaborn as sns

# Carrega o conjunto de dados 'tips'.
# Essa base contém informações relacionadas a contas e gorjetas em restaurantes.
dados = sns.load_dataset('tips')

# Mostra as 10 primeiras linhas da base completa.
dados.head(10)
```

### Separação correta entre X e Y

Em aprendizado de máquina supervisionado, normalmente separamos os dados em:

- `X`: entradas do modelo;
- `Y`: saída esperada.

Como queremos prever a gorjeta (`tip`), essa coluna deve sair de `X` e ser colocada em `Y`.

```python
# Remove a coluna 'tip' de X.
# axis=1 indica que estamos removendo uma coluna.
X = dados.drop('tip', axis=1)

# Mostra as 10 primeiras linhas das variáveis de entrada.
X.head(10)
```

```python
# Define Y como a variável de saída.
# Neste caso, Y contém apenas os valores de gorjeta.
Y = dados['tip']

# Mostra os 10 primeiros valores de Y.
Y.head(10)
```

```python
# Mostra o tamanho de X e Y.
# X deve ter 244 linhas e 6 colunas.
# Y deve ter 244 valores.
print('Formato de X:', X.shape)
print('Formato de Y:', Y.shape)
```

### O que ocorre se usar `X = dados`?

Se fizermos `X = dados`, a variável `X` passa a conter a base completa, inclusive a coluna `tip`.

Isso é inadequado quando `tip` é a variável que queremos prever, porque o modelo teria acesso à resposta durante o treinamento. Esse problema é conhecido como **vazamento de dados** (*data leakage*).

```python
# Exemplo propositalmente inadequado para fins didáticos:
# Aqui X recebe a base completa, incluindo a coluna 'tip'.
X_errado = dados

# Mostra as 10 primeiras linhas.
X_errado.head(10)
```

```python
# Comparando os formatos:
print('Formato correto de X, sem a coluna tip:', X.shape)
print('Formato de X_errado, com a coluna tip:', X_errado.shape)
```

### Interpretação da Tarefa 01

A forma correta para preparar os dados é usar `X = dados.drop('tip', axis=1)` e `Y = dados['tip']`. Assim, `X` contém apenas as variáveis explicativas, enquanto `Y` contém a resposta que o modelo deverá aprender a prever.

## LAB 04A — Tarefa 02

Base de dados: `titanic`

Essa base contém informações sobre passageiros do Titanic.

Nesta tarefa, a variável de saída é:

- `survived`: indica se o passageiro sobreviveu ou não.

Valores possíveis:

- `0`: não sobreviveu;
- `1`: sobreviveu.

```python
# LAB 04A 
# TAREFA 02

# Importa a biblioteca seaborn.
import seaborn as sns

# Carrega o conjunto de dados do Titanic usando a biblioteca seaborn.
titanic = sns.load_dataset('titanic')

# Mostra as 10 primeiras linhas da base Titanic.
titanic.head(10)
```

### Separação entre X e Y na base Titanic

Neste caso, queremos prever a variável `survived`. Portanto:

- `X` recebe todas as colunas, exceto `survived`;
- `Y` recebe apenas a coluna `survived`.

```python
# Remove a coluna 'survived' da base Titanic.
# Essa coluna é a resposta que queremos prever, por isso não deve ficar dentro de X.
X = titanic.drop('survived', axis=1)

# Define Y como a variável de saída.
Y = titanic['survived']
```

```python
# Mostra o tamanho da matriz X.
# Para a base Titanic do seaborn, normalmente X terá 891 linhas e 14 colunas.
X.shape
```

```python
# Mostra o tamanho de Y.
# Y terá 891 valores, um para cada passageiro.
Y.shape
```

```python
# Mostra as primeiras 10 linhas de X.
X.head(10)
```

```python
# Mostra os primeiros 10 valores de Y.
Y.head(10)
```

```python
# Impressão organizada dos formatos finais.
print('Formato de X:', X.shape)
print('Formato de Y:', Y.shape)
```

### Interpretação da Tarefa 02

A base `titanic` foi organizada para um problema de **classificação supervisionada**. A variável `survived` representa a classe que se deseja prever. As demais variáveis ficam em `X` e podem ser usadas como informações de entrada para o modelo.

Esse procedimento evita que a resposta fique misturada com as variáveis explicativas.

## Conclusão geral

Nas duas tarefas, o procedimento principal foi separar os dados em entrada e saída:

- Na base `tips`, a saída foi `tip`, caracterizando um problema de **regressão**, pois a gorjeta é um valor numérico.
- Na base `titanic`, a saída foi `survived`, caracterizando um problema de **classificação**, pois o resultado é uma classe: sobreviveu ou não sobreviveu.

Essa separação entre `X` e `Y` é uma etapa fundamental antes de treinar modelos de aprendizado de máquina.
