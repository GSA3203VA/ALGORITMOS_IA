```python
from bs4 import BeautifulSoup
import html
import re

arquivo_html = "Lab02Tarefa03.html"
arquivo_saida = "codigo_python_extraido.py"

with open(arquivo_html, "r", encoding="utf-8") as f:
    conteudo = f.read()

soup = BeautifulSoup(conteudo, "html.parser")

blocos = soup.find_all("div", class_=re.compile("highlight"))

codigos = []

for bloco in blocos:
    texto = bloco.get_text("\n", strip=False)
    texto = html.unescape(texto).strip()
    
    if any(p in texto for p in ["import ", "pd.", "plt.", "np.", "for ", "if ", "="]):
        codigos.append(texto)

with open(arquivo_saida, "w", encoding="utf-8") as f:
    for i, codigo in enumerate(codigos, start=1):
        f.write(f"# ===== CÓDIGO {i} =====\n")
        f.write(codigo)
        f.write("\n\n")

print("Extração concluída.")
print("Total de blocos encontrados:", len(codigos))
```
