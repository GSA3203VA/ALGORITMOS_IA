## Tarefa 01 — k-means com 2 variáveis

Neste notebook, aplicamos o algoritmo **k-means** à base `DadosWH.csv`, usando apenas duas variáveis:

- `Life_Ladder`
- `Perceptions_of_corruption`

O objetivo é agrupar os países de 2019 em clusters com características semelhantes.


Celula 1 - Importação das bibliotecas

Nesta etapa, importamos:

- `pandas` para leitura e manipulação da tabela de dados
- `matplotlib.pyplot` para gerar o gráfico
- `KMeans` do `sklearn` para aplicar o algoritmo de agrupamento


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Celula 2 - Leitura do arquivo CSV

Aqui lemos a base `DadosWH.csv` e armazenamos no DataFrame chamado `dados`.

O comando `head()` mostra as primeiras linhas para conferir se a leitura ocorreu corretamente.


```python
# Lê o arquivo CSV contendo os dados
dados = pd.read_csv('DadosWH.csv')

# Mostra as 5 primeiras linhas da base
dados.head()
```

Celula 3 - Filtrando apenas o ano de 2019

Como a atividade trabalha com os países em 2019, filtramos a coluna `year` e criamos um novo DataFrame chamado `dados2019`.

O uso de `.copy()` é recomendado para evitar conflitos ao editar a nova tabela depois.


```python
# Filtra somente as linhas cujo ano é 2019
dados2019 = dados[dados['year'] == 2019].copy()

# Mostra as primeiras linhas do recorte de 2019
dados2019.head()
```

## 4. Seleção das duas variáveis

Agora selecionamos apenas as duas colunas pedidas para o agrupamento:

- `Life_Ladder`: representa o nível de satisfação com a vida
- `Perceptions_of_corruption`: representa a percepção de corrupção

Essas duas colunas serão armazenadas na variável `Xs`.


```python
# Lista com os nomes das duas variáveis escolhidas
dimensoes = ['Life_Ladder', 'Perceptions_of_corruption']

# Cria uma nova tabela apenas com essas duas colunas
Xs = dados2019[dimensoes]

# Exibe as primeiras linhas das variáveis selecionadas
Xs.head()
```

## 5. Criação e treinamento do modelo k-means

Nesta etapa:

- `n_clusters=6` define que o algoritmo irá formar 6 grupos
- `random_state=42` garante reprodutibilidade dos resultados
- `n_init=10` faz o algoritmo testar 10 inicializações diferentes e escolher a melhor

Depois, `fit(Xs)` treina o modelo com as duas variáveis selecionadas.


```python
# Cria o modelo k-means com 6 clusters
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)

# Treina o modelo com os dados selecionados
kmeans.fit(Xs)
```

## 6. Salvando os clusters na tabela

Após o treinamento, o algoritmo atribui um número de cluster para cada país.

Esses rótulos ficam em `kmeans.labels_` e são salvos em uma nova coluna chamada `cluster`.


```python
# Adiciona ao DataFrame uma nova coluna com o cluster de cada país
dados2019['cluster'] = kmeans.labels_

# Mostra o nome do país e o cluster correspondente
dados2019[['Name', 'cluster']].head(10)
```

## 7. Exibindo os países pertencentes a cada cluster

O laço `for` percorre os clusters de 0 até 5 e mostra os países que pertencem a cada grupo.


```python
# Percorre cada cluster e mostra os países agrupados nele
for i in range(6):
    paises_cluster = dados2019[dados2019['cluster'] == i]['Name'].values
    print(f'\nCluster {i}:')
    print(paises_cluster)
```

## 8. Visualização gráfica dos clusters

Neste gráfico:

- cada ponto representa um país
- a cor indica o cluster ao qual ele pertence
- os centróides são marcados com `X` vermelho

Os centróides são os pontos centrais de cada cluster.


```python
# Cria uma figura com tamanho definido
plt.figure(figsize=(10, 6))

# Plota os países usando as duas variáveis escolhidas
plt.scatter(
    Xs['Life_Ladder'],                 # eixo x
    Xs['Perceptions_of_corruption'],  # eixo y
    c=kmeans.labels_,                 # cor de cada ponto conforme o cluster
    cmap='viridis',                   # mapa de cores
    s=60                              # tamanho dos pontos
)

# Obtém as coordenadas dos centróides
centroides = kmeans.cluster_centers_

# Plota os centróides com marcador em forma de X
plt.scatter(
    centroides[:, 0],
    centroides[:, 1],
    marker='X',
    s=250,
    color='red',
    label='Centróides'
)

# Título e rótulos dos eixos
plt.title('K-means com 2 variáveis')
plt.xlabel('Life_Ladder')
plt.ylabel('Perceptions_of_corruption')

# Mostra a legenda e a grade
plt.legend()
plt.grid(True)

# Exibe o gráfico
plt.show()
```

## 9. Conclusão

Nesta tarefa, o algoritmo k-means foi aplicado à base `DadosWH.csv`, considerando apenas os países do ano de 2019. O agrupamento foi feito com base em duas variáveis: `Life_Ladder` e `Perceptions_of_corruption`. Ao final, foi possível identificar grupos de países com características semelhantes nessas duas dimensões e visualizar graficamente a posição de cada cluster e de seus respectivos centróides.
