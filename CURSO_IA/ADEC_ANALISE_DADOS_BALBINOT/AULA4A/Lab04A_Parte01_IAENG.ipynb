```
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```
enc = LabelEncoder()
dados["sex"] = enc.fit_transform(dados['sex'].astype('str'))
dados["smoker"] = enc.fit_transform(dados['smoker'].astype('str'))
dados["day"] = enc.fit_transform(dados['day'].astype('str'))
dados["time"] = enc.fit_transform(dados['time'].astype('str'))
```

```
dados.head()
```

```
#
```

```
dados_normalizados = (dados - dados.min())/(dados.max()-dados.min())
dados_normalizados.head(10)
```

```
dados_padronizados = (dados - dados.mean())/dados.std()
dados_padronizados.head(10)
```
