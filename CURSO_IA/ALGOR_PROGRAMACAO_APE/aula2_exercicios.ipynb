Exercicios da Aula 2 convertidos para Jupyter.


```python
"""Exercicios da Aula 2 - Algoritmos e Programacao."""

import math
import random


def inverter_lista(valores):
    """Retorna uma nova lista com os valores em ordem inversa."""
    invertida = []
    indice = len(valores) - 1

    while indice >= 0:
        invertida.append(valores[indice])
        indice -= 1

    return invertida


def pi_monte_carlo(pontos):
    """Estima Pi sorteando pontos no quadrado unitario."""
    pontos_sorteados = 0
    pontos_dentro = 0

    while pontos_sorteados < pontos:
        x = random.random()
        y = random.random()

        if x**2 + y**2 < 1:
            pontos_dentro += 1

        pontos_sorteados += 1

    return 4 * pontos_dentro / pontos_sorteados


def media(valores):
    return sum(valores) / len(valores)


def desvio_padrao_populacional(valores):
    m = media(valores)
    soma_quadrados = 0

    for valor in valores:
        soma_quadrados += (valor - m) ** 2

    return math.sqrt(soma_quadrados / len(valores))


def rodadas_pi_monte_carlo(rodadas, pontos_por_rodada):
    resultados = []

    for _ in range(rodadas):
        resultados.append(pi_monte_carlo(pontos_por_rodada))

    return {
        "resultados": resultados,
        "media": media(resultados),
        "desvio": desvio_padrao_populacional(resultados),
    }


def bubble_sort(valores):
    ordenada = valores.copy()
    tamanho = len(ordenada)

    for i in range(tamanho):
        for j in range(0, tamanho - i - 1):
            if ordenada[j] > ordenada[j + 1]:
                ordenada[j], ordenada[j + 1] = ordenada[j + 1], ordenada[j]

    return ordenada


def insertion_sort(valores):
    ordenada = valores.copy()

    for i in range(1, len(ordenada)):
        chave = ordenada[i]
        j = i - 1

        while j >= 0 and ordenada[j] > chave:
            ordenada[j + 1] = ordenada[j]
            j -= 1

        ordenada[j + 1] = chave

    return ordenada
```

## Exemplos de execucao


```python
numeros = [42, 7, 91, 13, 58, 24, 76, 3, 65, 30]
print("Lista original:", numeros)
print("Lista invertida:", inverter_lista(numeros))
print("Bubble Sort:", bubble_sort(numeros))
print("Insertion Sort:", insertion_sort(numeros))

estatisticas = rodadas_pi_monte_carlo(rodadas=10, pontos_por_rodada=10000)
print("Estimativas de Pi:", estatisticas["resultados"])
print("Media:", estatisticas["media"])
print("Desvio padrao:", estatisticas["desvio"])
```
