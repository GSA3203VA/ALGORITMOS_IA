# Aula 3 - Recursividade

Fibonacci iterativo e recursivo em notebook.


```python
"""Exercicios da Aula 3 - recursividade."""


def fibonacci_recursivo(n):
    if n < 0:
        raise ValueError("n deve ser maior ou igual a zero")

    if n <= 1:
        return n

    return fibonacci_recursivo(n - 1) + fibonacci_recursivo(n - 2)


def fibonacci_iterativo(n):
    if n < 0:
        raise ValueError("n deve ser maior ou igual a zero")

    a, b = 0, 1

    for _ in range(n):
        a, b = b, a + b

    return a


def sequencia_fibonacci(quantidade, funcao=fibonacci_iterativo):
    valores = []

    for n in range(quantidade):
        valores.append(funcao(n))

    return valores
```

## Exemplos de execucao


```python
quantidade = 10
print("Fibonacci iterativo:", sequencia_fibonacci(quantidade))
print("Fibonacci recursivo:", sequencia_fibonacci(quantidade, fibonacci_recursivo))
```
