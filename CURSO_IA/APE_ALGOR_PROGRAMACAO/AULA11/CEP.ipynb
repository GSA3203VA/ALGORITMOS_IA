```python
import requests, json
# Requisição GET simples: API ViaCEP (sem autenticação)
url = "https://viacep.com.br/ws/94930060/json/"
resposta = requests.get(url)


# Inspecionar status ANTES de parsear
print(resposta.status_code) # 200 = OK
print(resposta.headers["Content-Type"])
if resposta.status_code == 200:
 dados = resposta.json()
 print("Cidade:", dados["localidade"])
 print("Estado:", dados["uf"])
 print("Cód. IBGE:", dados["ibge"])
else:
 print(f"Erro HTTP: {resposta.status_code}")
```
